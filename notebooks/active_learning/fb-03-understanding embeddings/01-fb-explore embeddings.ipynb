```python
start_type='Gene'
end_type='Probeset'
```

```python
paths = redis.meta_paths(start_type, end_type)
```

```python
representations = np.array([p.get_representation('embedding') for p in paths])
```

```python
names = [str(p) for p in paths]
```

```python
X_embedded = TSNE(n_components=2,perplexity=6,learning_rate=20).fit_transform(representations)
```

```python
def includes_name(compare_func):
    contain_transcript = [i for i, p in enumerate(names) if compare_func(p)]
    no_transcript = [i for i, p in enumerate(names) if not compare_func(p)]
    plt.scatter(X_embedded[contain_transcript,0],X_embedded[contain_transcript,1])
    plt.scatter(X_embedded[no_transcript,0],X_embedded[no_transcript,1])
    plt.show()
```

```python
includes_name(lambda n: 'Transcript' in n)
```

```python
includes_name(lambda n: 'MAPS' in n)
```

```python
includes_name(lambda n: 'NEXT' in n)
```

```python
includes_name(lambda n: len(n) > 70)
```
